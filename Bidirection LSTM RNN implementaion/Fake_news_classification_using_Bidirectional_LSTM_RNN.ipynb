```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df = df.dropna()
```

```python
X = df.drop('label',axis = 1)
```

```python
y = df['label']
```

```python
y.value_counts()
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
```

```python
message = X.copy()
```

```python
message
```

```python
message.reset_index(inplace = True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

corpus = []
for i in range(len(message)):
    review = re.sub('[^a-zA-Z]', ' ', message['title'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stop_words]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
voc_size = 5000
onehot_repr = [one_hot(words,voc_size) for words in corpus]
onehot_repr
```

```python
sent_lengths = 20
embedded_docs = pad_sequences(onehot_repr,padding = 'pre',maxlen = sent_lengths)
embedded_docs
```

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length = sent_lengths))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(loss='binary_crossentrophy',optimizer = 'adam',metrics=['accuracy'])
print(model.summary())
```

```python
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size = 0.33,random_state = 42)
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=20, batch_size=32)
```

```python
y_pred = model.predict(X_test)
y_pred = np.where(y_pred >= 0.5 , 1,0)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred_rf = classifier.predict(X_test)
y_pred_rf = np.where(y_pred_rf >= 0.5 , 1,0)
accuracy_score(y_test,y_pred_rf)
```

```python
report = classification_report(y_test, y_pred_rf, target_names=['False', 'True'])
print(report)
```
